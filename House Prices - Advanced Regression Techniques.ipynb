## Data Preprocessing Steps

### Handling null values

```python
# checking null values for columns
df.isnull().sum()
```

```python
# filtering columns having null values

missing_values = df.isnull().sum()

# Calculate the percentage of missing values per column
missing_percentage = (missing_values / len(df)) * 100

# Filter columns that have missing values
columns_with_missing_values = missing_values[missing_values > 0]

# Print the columns that have missing values along with their counts and percentages
print("Columns with missing values and thier percentages:")
for col in columns_with_missing_values.index:
    count = missing_values[col]
    percentage = missing_percentage[col]
    print(f"{col}: {count}  ({percentage:.2f}% missing)") 
```

```python
df = df.drop(columns=['Alley','FireplaceQu','PoolQC','Fence','MiscFeature'])
```

```python
# removing column Id as there is no such importance
df = df.drop(columns=['Id'])
```

### Replacing null values wrt missing values of columns

#### 1. LotFrontage

```python
df['LotFrontage'].unique()
```

```python
sns.histplot(df['LotFrontage'],kde=True)
# In terms of filling null values median imputation will work
```

```python
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].median())
```

#### 2. MasVnrType

```python
df['MasVnrType'].unique()
```

```python
df['MasVnrType'] = df['MasVnrType'].fillna(df['MasVnrType'].mode()[0])
# As MasVnrType is categorical thats why used mode
```
